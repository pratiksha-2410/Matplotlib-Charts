# Line Plots

```python
import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [10,20,25,30]

plt.plot(x,y , color='red',marker='o',markersize=7)
plt.title('Line Plot')
plt.xlabel('Roll No')
plt.ylabel('Marks')
plt.legend(['Data'])
plt.tight_layout()
plt.grid(color='green',linestyle='--')
plt.savefig('Line_plot.png')
plt.show()
```

```python
# Monthly Sales Line Plot

Months = ['Jan','Feb','March','April']
Sales = [100,150,130,170]

plt.figure(figsize=(8,5))
plt.plot(Months,Sales , color='blue',marker='^')
plt.title('Monthly Sales')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.grid(color='grey',linestyle='-.')
plt.tight_layout()
plt.savefig('Monthly_Sales_Line_Plot.png')
plt.show()
```

```python
# Product Comparision Line Plot

Months = ['Jan','Feb','Mar','Apr','May','Jun']

Product_A_Sales = [100,120,150,130,170,200]
Product_B_Sales = [80,120,180,220,240,260]

plt.figure(figsize=(8,6))
plt.plot(Months,Product_A_Sales,color='blue',marker='o',markersize=6)
plt.plot(Months,Product_B_Sales,color='orange',marker='o',markersize=6)
plt.title("Product Sales Comparison")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.legend(['Product A','Product B'])
plt.grid(color='green',linestyle='--',alpha=0.7) 
plt.tight_layout()
plt.savefig('Product Sales Comparison Line plot.png')    
plt.show()
```

```python
import numpy as np

x = np.arange(1, 6)
y = [10, 20, 15, 25, 30]

plt.figure(figsize=(8,5))

plt.plot(x, y, marker='o', label="Trend")

# Highlight area
plt.fill_between(x, y, alpha=0.2)

# Annotation
plt.annotate("Peak", xy=(5,30), xytext=(4,30),
             arrowprops=dict(facecolor='black'))

plt.title("Advanced Analysis")
plt.legend()
plt.grid()

plt.show()
```

```python
# Profit & Loss Analysis

import numpy as np 
Days = np.arange(1,11)
Profit = [10,-5,15,-10,20,-15,25,-20,30,-25]

plt.plot(Days, Profit, marker='o', markersize=6,color='purple')
plt.title("Profit & Loss Analysis")
plt.xlabel("Days")
plt.ylabel("Profit/Loss")
plt.axhline(0,color='black',linewidth=0.5,linestyle='--')
plt.grid(color='orange',linestyle=':')
plt.legend(['Profit/Loss'])
plt.tight_layout()

# Highlight Profit Loss Area
plt.fill_between(Days,Profit, where=(np.array(Profit)>0),alpha=0.8,color='green')
plt.fill_between(Days,Profit,where=(np.array(Profit)<0),alpha=0.8,color='red')

# Annotation For Max Profit
plt.annotate('Max Profit',xy=(9,30),xytext=(7,30),arrowprops=dict(facecolor='black'))

# Save The Plot
plt.savefig('Profit loss Analysis Line Plot.png')
plt.show()
```

```python
# Sales & Profit Analysis with Subplots

import numpy as np 

Days = np.arange(1,11)
Sales = [100,150,130,170,200,220,250,270,300,320]
Profit = [10,-5,15,-10,20,-15,25,-20,30,-25]

fig,ax = plt.subplots(1,2,figsize=(10,5))

# Sales Plot 
ax[0].plot(Days,Sales,color='purple',marker='o')
ax[0].set_title('Sales Analysis')
ax[0].set_xlabel('Days')
ax[0].set_ylabel('Sales')
ax[0].grid(color='grey',linestyle='-.')

# Profit Plot

ax[1].plot(Days,Profit,color='orange',marker='^')
ax[1].set_xlabel('Days')
ax[1].set_ylabel('Profit')
ax[1].grid(color='grey',linestyle='-.')
ax[1].set_title('Profit Analysis')

fig.suptitle('Sales & Profit Analysis')
plt.tight_layout()
plt.savefig('Sales & Profit Analysis Line Plot.png')
plt.show()
```

```python
# Line & Bar Plot Comibination

Months = ['Jan','Feb','Mar','Apr','May','Jun']

Shirts_Sales = [100,120,150,130,170,200]
pants_Sales = [80,110,140,120,160,190]

fig,ax = plt.subplots(1,2,figsize=(10,5))

# Line Plot

ax[0].plot(Months,Shirts_Sales,color='Purple',marker='o')
ax[0].set_title('Month Wise Shirt Sales')
ax[0].set_xlabel('Months')
ax[0].set_ylabel('Shirts Sale')
ax[0].grid(color='green',linestyle=':',alpha=1)

# Bar Plot

ax[1].bar(Months,pants_Sales,color='blue')
ax[1].set_title('Month Wise Pant Sale')
ax[1].set_xlabel('Months')
ax[1].set_ylabel('Pants Sale')


fig.suptitle("Month Wise Garment Sale Analysis")

plt.tight_layout()
plt.savefig('Month Wise Garment Sale Analysis')
plt.show()
```

```python
# Month Wise Fraud Count Analysis

Months = ['Jan','Feb','Mar','Apr','May','Jun']
Fraud_Count = [100,150,130,170,200,220]

plt.figure(figsize=(10,5))

plt.plot(Months,Fraud_Count,color='purple',marker='o',markersize='6')
plt.title('Month Wise Fraud Count Analysis')
plt.xlabel('Months')
plt.ylabel('Fraud_Count')
plt.grid(color='grey',linestyle='--')
plt.tight_layout()

# Highlight Area
plt.fill_between(Months,Fraud_Count,alpha=0.5,color='red')

# Annotation For Peak Fraud Count

plt.annotate('Max Fraud Count',xy=('Jun',220),xytext=('Apr',220),arrowprops=dict(facecolor='black'))

# Save The Plot 

plt.savefig('Month wise Fraud Count Analysis line plot.png')
plt.show()
```

# Bar Plots

```python
# Transaction-Type VS Fraud Count Analysis

Transaction_Type = ['Online','Debit_Card','Credit_Card','Cash','Bill_Payment','Withdrawal','Transfer']
Fraud_Count = [2000,1500,1800,500,1200,800,1000]

plt.figure(figsize=(8,6))

plt.bar(Transaction_Type,Fraud_Count,color='grey')
plt.title('Transaction-Type Wise Fraud Count Analysis')
plt.xlabel('Transaction_Type')
plt.ylabel('Fraud_Count')

# Annotation For Highest Fraud Count
plt.annotate('Max Fraud Count',xy=('Online',2000),xytext=('Debit_Card',2000),arrowprops=dict(facecolor='black'))
plt.tight_layout()
plt.savefig('Transaction Type Vs Fraud Count Analysis Bar Plot.png')
plt.show()
```
