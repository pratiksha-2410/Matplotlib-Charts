# Line Plots

```python
import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [10,20,25,30]

plt.plot(x,y , color='red',marker='o',markersize=7)
plt.title('Line Plot')
plt.xlabel('Roll No')
plt.ylabel('Marks')
plt.legend(['Data'])
plt.tight_layout()
plt.grid(color='green',linestyle='--')
plt.savefig('Line_plot.png')
plt.show()
```

```python
# Monthly Sales Line Plot

Months = ['Jan','Feb','March','April']
Sales = [100,150,130,170]

plt.figure(figsize=(8,5))
plt.plot(Months,Sales , color='blue',marker='^')
plt.title('Monthly Sales')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.grid(color='grey',linestyle='-.')
plt.tight_layout()
plt.savefig('Monthly_Sales_Line_Plot.png')
plt.show()
```

```python
# Product Comparision Line Plot

Months = ['Jan','Feb','Mar','Apr','May','Jun']

Product_A_Sales = [100,120,150,130,170,200]
Product_B_Sales = [80,120,180,220,240,260]

plt.figure(figsize=(8,6))
plt.plot(Months,Product_A_Sales,color='blue',marker='o',markersize=6)
plt.plot(Months,Product_B_Sales,color='orange',marker='o',markersize=6)
plt.title("Product Sales Comparison")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.legend(['Product A','Product B'])
plt.grid(color='green',linestyle='--',alpha=0.7) 
plt.tight_layout()
plt.savefig('Product Sales Comparison Line plot.png')    
plt.show()
```

```python
import numpy as np

x = np.arange(1, 6)
y = [10, 20, 15, 25, 30]

plt.figure(figsize=(8,5))

plt.plot(x, y, marker='o', label="Trend")

# Highlight area
plt.fill_between(x, y, alpha=0.2)

# Annotation
plt.annotate("Peak", xy=(5,30), xytext=(4,30),
             arrowprops=dict(facecolor='black'))

plt.title("Advanced Analysis")
plt.legend()
plt.grid()

plt.show()
```

```python
# Profit & Loss Analysis

import numpy as np 
Days = np.arange(1,11)
Profit = [10,-5,15,-10,20,-15,25,-20,30,-25]

plt.plot(Days, Profit, marker='o', markersize=6,color='purple')
plt.title("Profit & Loss Analysis")
plt.xlabel("Days")
plt.ylabel("Profit/Loss")
plt.axhline(0,color='black',linewidth=0.5,linestyle='--')
plt.grid(color='orange',linestyle=':')
plt.legend(['Profit/Loss'])
plt.tight_layout()

# Highlight Profit Loss Area
plt.fill_between(Days,Profit, where=(np.array(Profit)>0),alpha=0.8,color='green')
plt.fill_between(Days,Profit,where=(np.array(Profit)<0),alpha=0.8,color='red')

# Annotation For Max Profit
plt.annotate('Max Profit',xy=(9,30),xytext=(7,30),arrowprops=dict(facecolor='black'))

# Save The Plot
plt.savefig('Profit loss Analysis Line Plot.png')
plt.show()
```
