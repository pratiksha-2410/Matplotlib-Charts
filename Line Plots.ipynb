# Line Plots

```python
import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [10,20,25,30]

plt.plot(x,y , color='red',marker='o',markersize=7)
plt.title('Line Plot',fontweight='bold',pad=20)
plt.xlabel('Roll No',fontweight='bold',labelpad=20)
plt.ylabel('Marks',fontweight='bold',labelpad=20)
plt.legend(['Data'])
plt.tight_layout()
plt.grid(color='green',linestyle='--')
plt.savefig('Line_plot.png')
plt.show()
```

```python
# Monthly Sales Line Plot

Months = ['Jan','Feb','March','April']
Sales = [100,150,130,170]

plt.figure(figsize=(8,5))
plt.plot(Months,Sales , color='blue',marker='^')
plt.title('Monthly Sales',fontweight='bold',pad=20)
plt.xlabel('Months',fontweight='bold',labelpad=20)
plt.ylabel('Sales',fontweight='bold',labelpad=20)
plt.grid(color='grey',linestyle='-.')
plt.tight_layout()
plt.savefig('Monthly_Sales_Line_Plot.png')
plt.show()
```

```python
# Product Comparision Line Plot

Months = ['Jan','Feb','Mar','Apr','May','Jun']

Product_A_Sales = [100,120,150,130,170,200]
Product_B_Sales = [80,120,180,220,240,260]

plt.figure(figsize=(8,6))

plt.plot(Months,Product_A_Sales,color='blue',marker='o',markersize=6)
plt.plot(Months,Product_B_Sales,color='orange',marker='o',markersize=6)
plt.title("Product Sales Comparison",fontweight='bold',pad=20)
plt.xlabel("Months",fontweight='bold',labelpad=20)
plt.ylabel("Sales",fontweight='bold',labelpad=20)

plt.legend(['Product A','Product B'])
plt.grid(color='green',linestyle='--',alpha=0.7) 
plt.tight_layout()
plt.xticks(fontname='Arial')
plt.yticks(fontname='Arial')

# Save The Plot
plt.savefig('Product Sales Comparison Line plot.png')    
plt.show()
```

```python
import numpy as np

x = np.arange(1, 6)
y = [10, 20, 15, 25, 30]

plt.figure(figsize=(8,5))

plt.plot(x, y, marker='o', label="Trend")

# Highlight area
plt.fill_between(x, y, alpha=0.2)

# Annotation
plt.annotate("Peak", xy=(5,30), xytext=(4,30),
             arrowprops=dict(arrowstyle='->', facecolor='black'),fontweight='bold')

plt.title("Advanced Analysis",fontweight='bold',pad=20)
plt.legend()
plt.grid()

plt.show()
```

```python
# Profit & Loss Analysis

import numpy as np 
Days = np.arange(1,11)
Profit = [10,-5,15,-10,20,-15,25,-20,30,-25]

plt.plot(Days, Profit, marker='o', markersize=6,color='purple')
plt.title("Profit & Loss Analysis",fontweight='bold',pad=20)
plt.xlabel("Days",fontweight='bold',labelpad=20)
plt.ylabel("Profit/Loss",fontweight='bold',labelpad=20)
plt.axhline(0,color='black',linewidth=0.5,linestyle='--')
plt.grid(color='orange',linestyle=':')
plt.legend(['Profit/Loss'])
plt.tight_layout()

# Highlight Profit Loss Area
plt.fill_between(Days,Profit, where=(np.array(Profit)>0),alpha=0.8,color='green')
plt.fill_between(Days,Profit,where=(np.array(Profit)<0),alpha=0.8,color='red')

# Annotation For Max Profit
plt.annotate('Max Profit',xy=(9,30),xytext=(7,30),arrowprops=dict(arrowstyle='->', facecolor='black'),fontweight='bold')

# Save The Plot
plt.savefig('Profit loss Analysis Line Plot.png')
plt.show()
```

```python
# Sales & Profit Analysis with Subplots

import numpy as np 

Days = np.arange(1,11)
Sales = [100,150,130,170,200,220,250,270,300,320]
Profit = [10,-5,15,-10,20,-15,25,-20,30,-25]

fig,ax = plt.subplots(1,2,figsize=(10,5))

# Sales Plot 
ax[0].plot(Days,Sales,color='purple',marker='o')
ax[0].set_title('Sales Analysis',fontweight='bold',pad=20)
ax[0].set_xlabel('Days',fontweight='bold',labelpad=20)
ax[0].set_ylabel('Sales',fontweight='bold',labelpad=10)
ax[0].grid(color='grey',linestyle='-.')

# Profit Plot

ax[1].plot(Days,Profit,color='orange',marker='^')
ax[1].set_xlabel('Days',fontweight='bold',labelpad=20)
ax[1].set_ylabel('Profit',fontweight='bold',labelpad=5)
ax[1].grid(color='grey',linestyle='-.')
ax[1].set_title('Profit Analysis',fontweight='bold',pad=20)

fig.suptitle('Sales & Profit Analysis',fontweight='bold',color='orange')
plt.tight_layout()
plt.savefig('Sales & Profit Analysis Line Plot.png')
plt.show()
```

```python
# Line & Bar Plot Comibination

Months = ['Jan','Feb','Mar','Apr','May','Jun']

Shirts_Sales = [100,120,150,130,170,200]
pants_Sales = [80,110,140,120,160,190]

fig,ax = plt.subplots(1,2,figsize=(10,5))

# Line Plot

ax[0].plot(Months,Shirts_Sales,color='Purple',marker='o')
ax[0].set_title('Month Wise Shirt Sales',fontweight='bold',pad=20)
ax[0].set_xlabel('Months',fontweight='bold',labelpad=20)
ax[0].set_ylabel('Shirts Sale',fontweight='bold',labelpad=20)
ax[0].grid(color='green',linestyle=':',alpha=1)

# Bar Plot

ax[1].bar(Months,pants_Sales,color='blue')
ax[1].set_title('Month Wise Pant Sale',fontweight='bold',pad=20)
ax[1].set_xlabel('Months',fontweight='bold',labelpad=20)
ax[1].set_ylabel('Pants Sale',fontweight='bold',labelpad=20)


fig.suptitle("Month Wise Garment Sale Analysis",fontweight='bold',color='darkblue')

plt.tight_layout()
plt.savefig('Month Wise Garment Sale Analysis')
plt.show()
```

```python
# Month Wise Fraud Count Analysis

Months = ['Jan','Feb','Mar','Apr','May','Jun']
Fraud_Count = [100,150,130,170,200,220]

plt.figure(figsize=(10,5))

plt.plot(Months,Fraud_Count,color='purple',marker='o',markersize='6')
plt.title('Month Wise Fraud Count Analysis',fontweight='bold',pad=20,fontsize=14)
plt.xlabel('Months',fontweight='bold',labelpad=20,fontsize=12)
plt.ylabel('Fraud_Count',fontweight='bold',labelpad=20,fontsize=12)
plt.grid(color='grey',linestyle='--')
plt.tight_layout()

# Highlight Area
plt.fill_between(Months,Fraud_Count,alpha=0.5,color='red')

# Annotation For Peak Fraud Count

plt.annotate('Max Fraud Count',xy=('Jun',220),xytext=('Apr',220),arrowprops=dict(arrowstyle='->', facecolor='black'),fontweight='bold')

# Save The Plot 

plt.savefig('Month wise Fraud Count Analysis line plot.png')
plt.show()
```

```python
# Transaction-Type VS Fraud Count Analysis

Transaction_Type = ['Online','Debit_Card','Credit_Card','Cash','Bill_Payment','Withdrawal','Transfer']
Fraud_Count = [2000,1500,1800,500,1200,800,1000]

plt.figure(figsize=(8,6))

plt.bar(Transaction_Type,Fraud_Count,color='grey',alpha=1)
plt.title('Transaction-Type Wise Fraud Count Analysis',fontweight='bold',pad=20)
plt.xlabel('Transaction_Type',fontweight='bold',labelpad=20)
plt.ylabel('Fraud_Count',fontweight='bold',labelpad=20)

# Annotation For Highest Fraud Count
plt.annotate('Max Fraud Count',xy=('Online',2000),xytext=('Debit_Card',2000),arrowprops=dict(arrowstyle='->', facecolor='black'),fontweight='bold')
plt.tight_layout()
plt.savefig('Transaction Type Vs Fraud Count Analysis Bar Plot.png')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

Categories = ['Electronics','Clothing','Groceries','Furniture','Toys']

Jan_Sales = [5000,7000,6000,4000,3000]
March_Sales = [5500,7500,6500,4500,3500]

x = np.arange(len(Categories))
width = 0.35

plt.figure(figsize=(8,5))

plt.bar(x - width/2, Jan_Sales, width=width, color='blue', label='Jan_Sales',alpha=0.7)
plt.bar(x + width/2, March_Sales, width=width, color='orange', label='March_Sales')

plt.title('Category Wise Sales Analysis',fontweight='bold',fontsize=14,pad=20)
plt.xlabel('Categories',fontweight='bold',labelpad=20)
plt.ylabel('Sales',fontweight='bold',labelpad=20)

plt.xticks(x, Categories)
plt.legend()

# annotation
plt.annotate('Highest Sale', xy=(1,7500), xytext=(2,7500),
             arrowprops=dict(arrowstyle='->', facecolor='black'),fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
# Category Wise VOte For Political Parties Analysis

Parties = ['BJP','Congress','AAP','SP','BSP']
Male_Votes = [5000,3000,2000,1500,1000]
Female_Votes = [4000,3500,2500,2000,1500]

x = np.arange(len(Parties))
width = 0.35

plt.figure(figsize=(8,5))

plt.bar(x - width/2,Male_Votes,width=width,color='green',alpha=0.9)
plt.bar(x + width/2,Female_Votes,width=width,color='orange',alpha=0.9)
plt.title('Category Wise Vote For Political Parties Analysis',fontweight='bold',pad=20)
plt.xlabel('Parties',fontweight='bold',labelpad=20,fontsize=12)
plt.ylabel('Votes',fontweight='bold',labelpad=20,fontsize=12)
plt.legend(['Male_Votes', 'Female_Votes'])

# Annotation For Highest Votes
plt.annotate('Highest Votes',xy=(0,5000),xytext=(1,5000),arrowprops=dict(facecolor='black'),fontweight='bold')

plt.xticks(x,Parties)
plt.savefig('Category Wise Vote For Political Parties Analysis Bar Plot.png')
plt.show()
```

```python
# Social Media Rating Analysis Horizontal Bar Plot

Rating = [45,40,83,75,60]
Social_Platforms = ['Whatsapp','Twitter','Instagram','Facebook','Snapchat']

plt.figure(figsize=(8,5))

plt.barh(Social_Platforms,Rating,color='Blue',height=0.5)
plt.title('Social Media Rating Analysis',fontweight='bold',pad=20)
plt.ylabel('Social Media',color='red',fontweight='bold',fontsize=14,labelpad=20)
plt.xlabel('Rating(%)',color='red',fontweight='bold',fontsize=14,labelpad=20)

# Labeling Bars
plt.yticks(color='green',fontname='Arial',fontsize=12,rotation=0)
plt.xticks(color='green',fontname='Arial',fontsize=12)

plt.tight_layout()
# Save the plot 
plt.savefig('Social Media Rating Analysis')
plt.show()
```
